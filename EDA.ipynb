```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['TotalPay'], bins=40, kde=True, stat="percent")
plt.title('Distribución porcentual de Salarios Totales')
plt.xlabel('TotalPay')
plt.ylabel('Porcentaje (%)')
plt.show()
```

```python
df.groupby('Year')['TotalPay'].mean().plot(kind='bar')
plt.title('Salario Promedio por Año')
plt.ylabel('Promedio de TotalPay')
plt.show()
```

```python
top_jobs = df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False).head(10)
print(top_jobs)
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Top 10 cargos mejor pagados')
plt.xlabel('Salario promedio')
plt.show()
```

```python
cols = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']

print("ANTES DE LA LIMPIEZA:")
print(df[cols].dtypes)
print("\nValores únicos no numéricos encontrados:")
for c in cols:
    no_numericos = df[c].apply(lambda x: isinstance(x, str)).sum()
    if no_numericos > 0:
        print(f" - {c}: {no_numericos} valores tipo texto")

for c in cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')  # Los errores se vuelven NaN

print("\nDESPUÉS DE LA LIMPIEZA:")
print(df[cols].dtypes)

df[cols] = df[cols].applymap(lambda x: np.nan if x < 0 else x)

print("\nValores nulos por columna después de limpiar:")
print(df[cols].isnull().sum())

print("\nVista rápida de las primeras filas:")
print(df[cols].head())
```

```python
componentes = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
df[componentes].mean().plot(kind='bar', color='skyblue')
plt.title('Promedio de componentes del salario')
plt.ylabel('Monto promedio ($)')
plt.show()
```

```python
outliers = df[df['TotalPay'] > df['TotalPay'].quantile(0.99)]
print(outliers[['EmployeeName','JobTitle','TotalPay']])
```

## Conclusiones

- La mayoría de los salarios se concentran entre 50 000 y 100 000 USD.
- Los pagos por horas extra representan un 15 % promedio del total.
- Existen algunos outliers con salarios superiores a 300 000 USD, posiblemente directivos.
- Los salarios crecieron ligeramente entre 2011 y 2014, pero sin gran variación.
-El análisis de los salarios en la ciudad de San Francisco evidencia una marcada concentración de los ingresos más altos en los cargos directivos. Como se observa en el gráfico de los Top 10 cargos mejor pagados, los puestos con mayores remuneraciones corresponden principalmente a funciones de dirección y gestión dentro de instituciones públicas, tales como el General Manager – Metropolitan Transit Authority y el Chief Investment Officer.
